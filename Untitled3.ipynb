(a) Retrieve a list of all movie-titles, in which ANY actor of the movie ”The Matrix”
has acted in. In the output, the movie-title should show up only once, not multiple
times!

```python
from py2neo import Graph, Node, Relationship
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

graph = Graph("bolt://localhost:7687", auth=("neo4j", "123"))
query = "MATCH (m1:Movie{title:'The Matrix'})<-[r1:ACTED_IN]-(p1:Person) \
MATCH (p2:Person)-[r2:ACTED_IN]->(m2:Movie) WHERE p2.name = p1.name RETURN DISTINCT m2.title"
result = graph.run(query)
#result.to_table()
print(result.to_table())
```

(b) Retrieve a table of all actor-names and the number of movies, in which they
acted. The columns in the output should be called ”Actors-Name” and ”Number of
movies”.

```python
graph = Graph("bolt://localhost:7687", auth=("neo4j", "123"))
query = "MATCH (p:Person)-[r:ACTED_IN]-(m:Movie) \
        RETURN p.name AS `Actors-Name`,\
        COUNT(DISTINCT m) AS `Number of movies` \
        ORDER BY `Number of movies` DESC"
result = graph.run(query)
print(result.to_table())
```

(c) Retrieve a connection between the actor ”Tom Hanks” and the movie ”The Matrix”.

```python
import networkx as nx

graph = Graph("bolt://localhost:7687", auth=("neo4j", "123"))
query = "MATCH (p:Person { name: 'Tom Hanks' }),(m:Movie { title: 'The Matrix' }),\
        shortest_Path = shortestPath((p)-[*]-(m))\
        RETURN shortest_Path"
#result = graph.run(query)
#print(result.to_table())

df = pd.DataFrame(graph.run(query))
print(df)
nx.draw(df)
```

(d) Retrieve the minimal, the maximal and the average rating of all movies AND the
number of users who rated that movie.

```python
MATCH (m:Movie)<-[r:RATED]-(u:User)
RETURN m.title AS movie, count(u) AS Number_Of_User, avg(r.rating) AS Rating, min(r.rating)  AS Min, max(r.rating) AS Max ORDER BY  avg(r.rating) DESC
```
